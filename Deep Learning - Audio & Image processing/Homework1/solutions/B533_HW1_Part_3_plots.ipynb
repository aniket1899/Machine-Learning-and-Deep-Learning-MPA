```python
import numpy as np
import pandas as pd
# import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
# from tensorflow.keras.callbacks import CSVLogger

%matplotlib inline 

# %tensorflow_version 2.x
```

```python
xav_nodrop = pd.read_csv('Part 2 accuracies and losses/xav_nodrop_results.csv')
xav_drop = pd.read_csv('Part 2 accuracies and losses/xav_drop_results.csv')

he_nodrop = pd.read_csv('Part 2 accuracies and losses/he_nodrop_results.csv')
he_drop = pd.read_csv('Part 2 accuracies and losses/he_drop_results.csv')
```

```python
xav = pd.DataFrame(xav_nodrop[['epoch', 'loss', 'val_loss']])
xav.columns = ['epoch', 'No Dropout: Train loss', 'No Dropout: Test loss']
xav['Dropout: Train loss'] = xav_drop['loss']
xav['Dropout: Test loss'] = xav_drop['val_loss']
```

```python
# xav
```

#### Plotting train and test loss for model with the following attributes:
1. Activation function: the logistic sigmoid function; initialization: Xavier initializer; no
dropout
2. Activation function: the logistic sigmoid function; initialization: Xavier initializer; with
dropout rate: 0.2 for the â€‚rst layer and 0.5 for the other hidden layers

```python
fig = plt.figure(figsize=(15,10))
ax = sns.lineplot(data=xav, x="epoch", y="No Dropout: Train loss", label="No Dropout: Train loss")
sns.lineplot(data=xav, x="epoch", y="No Dropout: Test loss", label="No Dropout: Test loss")
sns.lineplot(data=xav, x="epoch", y="Dropout: Train loss", label="Dropout: Train loss")
sns.lineplot(data=xav, x="epoch", y="Dropout: Test loss", label="Dropout: Test loss")
# fig.legend(, label="")
plt.title("Cross Entropy Loss comparison (1024x5 Logistic)", fontsize=14)
# ax.xlabel("")
ax.set(ylabel='Cross Entropy Error', xlabel='Epochs')
plt.show()
```

```python
he = pd.DataFrame(he_nodrop[['epoch', 'loss', 'val_loss']])
he.columns = ['epoch', 'No Dropout: Train loss', 'No Dropout: Test loss']
# he['No Dropout: Train loss'] = he['No Dropout: Train loss']
he['Dropout: Train loss'] = he_drop['loss']
he['Dropout: Test loss'] = he_drop['val_loss']
```

#### Plotting train and test loss for model with the following attributes:
1. Activation function: ReLU; initialization: Kaiming He's initializer; no dropout
2. Activation function: ReLU; initialization: Kaiming He's initializer; with dropout rate:
0.2 for the â€‚rst layer and 0.5 for the other hidden layers

```python
fig = plt.figure(figsize=(15,10))
ax = sns.lineplot(data=he, x="epoch", y="No Dropout: Train loss", label="No Dropout: Train loss")
sns.lineplot(data=he, x="epoch", y="No Dropout: Test loss", label="No Dropout: Test loss")
sns.lineplot(data=he, x="epoch", y="Dropout: Train loss", label="Dropout: Train loss")
sns.lineplot(data=he, x="epoch", y="Dropout: Test loss", label="Dropout: Test loss")
# fig.legend(, label="")
plt.title("Cross Entropy Loss comparison (1024x5 ReLU)", fontsize=14)
# ax.xlabel("")
ax.set(ylabel='Cross Entropy Error', xlabel='Epochs')
plt.show()
```
