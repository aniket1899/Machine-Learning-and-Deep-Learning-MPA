```python
a_sgd = pd.read_csv('Part 2 accuracies and losses/a_sgd_results.csv')
a_adam = pd.read_csv('Part 2 accuracies and losses/a_adam_results.csv')

b_sgd = pd.read_csv('Part 2 accuracies and losses/b_sgd_results.csv')
b_adam = pd.read_csv('Part 2 accuracies and losses/b_adam_results.csv')

c_sgd = pd.read_csv('Part 2 accuracies and losses/c_sgd_results.csv')
c_adam = pd.read_csv('Part 2 accuracies and losses/c_adam_results.csv')

d_sgd = pd.read_csv('Part 2 accuracies and losses/d_sgd_results.csv')
d_adam = pd.read_csv('Part 2 accuracies and losses/d_adam_results.csv')

e_sgd = pd.read_csv('Part 2 accuracies and losses/e_sgd_results.csv')
e_adam = pd.read_csv('Part 2 accuracies and losses/e_adam_results.csv')

```

```python
# e_sgd.head()
```

```python
sgd_accuracies = pd.DataFrame(a_sgd[['epoch', 'val_accuracy']])
sgd_accuracies.columns = ['epoch', 'Logistic: Normal']
sgd_accuracies['Logistic: Normal'] = sgd_accuracies['Logistic: Normal'] *100
sgd_accuracies['Logistic: Xavier'] = b_sgd['val_accuracy']*100
sgd_accuracies['ReLU: Normal'] = c_sgd['val_accuracy']*100
sgd_accuracies['ReLU: Xavier'] = d_sgd['val_accuracy']*100
sgd_accuracies['ReLU: He'] = e_sgd['val_accuracy']*100
# sgd_accuracies.head()
```

#### Plotting testing accuracy for the five models with Optimizer = SGD


```python
fig = plt.figure(figsize=(15,10))
ax =sns.lineplot(data=sgd_accuracies, x="epoch", y="Logistic: Normal", label="Logistic: Normal")
sns.lineplot(data=sgd_accuracies, x="epoch", y="Logistic: Xavier", label="Logistic: Xavier")
sns.lineplot(data=sgd_accuracies, x="epoch", y="ReLU: Normal", label="ReLU: Normal")
sns.lineplot(data=sgd_accuracies, x="epoch", y="ReLU: Xavier", label="ReLU: Xavier")
sns.lineplot(data=sgd_accuracies, x="epoch", y="ReLU: He", label="ReLU: He")
# fig.legend(, label="")
plt.title("Testing accuracy for different models (with optimizer = SGD)", fontsize=14)
# ax.xlabel("")
ax.set(ylabel='Test accuracy %', xlabel='Epochs')
plt.show()
```
