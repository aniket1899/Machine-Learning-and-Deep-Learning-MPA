```python
# load the data

s, sr=librosa.load("drive/My Drive/b533_hw2/b533_hw2_data/train_clean_male.wav", sr=None)
S=librosa.stft(s, n_fft=1024, hop_length=512)
sn, sr=librosa.load("drive/My Drive/b533_hw2/b533_hw2_data/train_dirty_male.wav", sr=None)
X = librosa.stft(sn, n_fft=1024, hop_length=512)
print("Shape of S:", S.shape)
print("Shape of X:", X.shape)
```

```python
# extracting magnitudes
Xmag = np.abs(X)
Smag = np.abs(S)
```

```python
# Smag.shape
```

#### Constructing the 1D CNN model

```python
model_1d = Sequential([
    # Input(shape=((5,1,1,513))),
    Input(shape=(513,1)),
    # Conv1D(64, 3, activation='relu', strides=2),
    # MaxPooling1D(),
    # Dropout(0.5),
    Conv1D(32, 5, activation='relu', strides=2),
    MaxPooling1D(),
    # Dropout(0.5),
    Conv1D(32, 3, activation='relu', strides=2),
    MaxPooling1D(),

    Flatten(),
    # Dense(1024, activation='relu'),

    Dense(513, activation='relu')
    
]
)

model_1d.compile(
    loss = 'mse',
    optimizer = Adam(learning_rate=0.001),
    metrics = ['Accuracy']
)

model_1d.summary()
```

```python
model_1d_fit = model_1d.fit(
    Xmag.T.reshape(2459, 513, 1),
    Smag.T,
    epochs=100, 
    # validation_split=0.20, 
    batch_size=15, 
    verbose=2)
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig = make_subplots(rows=1, cols=2, subplot_titles=("Training Loss", "Training Accuracy"))

fig.add_trace(
    # px.line(y=model_1d_fit.history['loss'], 
    #               title='Training Loss',
    #               name='Training Loss'),
    go.Scatter(y=model_1d_fit.history['loss'],
               line_color='red',
               name='Training Loss'
               ),
    row=1, col=1
)

fig.add_trace(
    # px.line(y=model_1d_fit.history['Accuracy'], 
    #               title='Training Accuracy',
    #               name='Training Accuracy'),

    go.Scatter(y=model_1d_fit.history['Accuracy'],
               line_color='blue',
               name='Training Accuracy'
               ),
    row=1, col=2
)
fig.update_yaxes(title_text="loss", row=1, col=1)
fig.update_yaxes(title_text="accuracy", row=1, col=2)
fig.update_layout(height=600, width=800, title_text="HW2:P2 Conv1D model metrics")
fig.show()
```
